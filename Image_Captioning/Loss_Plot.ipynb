```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('C:/Workspace/python/빅데이터 지능형서비스 개발 팀프로젝트/Final Project/Data/IC_Data/모델 학습내용(Keras TK).txt', sep = ' ', header = None)
df = df[[5, 8]].copy()
df.columns = ['loss(train)', 'loss(val)']
df.head()
```

```python
loss_plot = list(df['loss(train)'])
val_loss_plot = list(df['loss(val)'])
```

```python
plt.plot(loss_plot)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Keras_Tokenizer_Loss Plot')
plt.show()
```

```python
plt.plot(val_loss_plot)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Keras_Tokenizer_Loss Plot')
plt.show()
```
