```python
from hanspell import spell_checker
from tqdm.notebook import tqdm
import re

image_path_to_caption = collections.defaultdict(list)
# for val in annotations:
#   for i in range(len(val['caption_ko'])):
#     caption = f"start {val['caption_ko'][i]} end"
#     image_path = val['file_path']
#     image_path_to_caption[image_path].append(caption)

for index, val in tqdm(enumerate(df['캡셔닝_데이터'])):
    caption = f"start {val} end"
    image_path = df['사진경로'][index]
    image_path_to_caption[image_path].append(caption)
```

```python
image_paths = list(image_path_to_caption.keys())
random.shuffle(image_paths)

# Select the first 6000 image_paths from the shuffled set.
# Approximately each image id has 5 captions associated with it, so that will
# lead to 30,000 examples.
train_image_paths = image_paths
print(len(train_image_paths), train_image_paths[0], sep = ' ')
```

```python
train_captions = []
img_name_vector = []

for image_path in train_image_paths:
  caption_list = image_path_to_caption[image_path]
  train_captions.extend(caption_list)
  img_name_vector.extend([image_path] * len(caption_list))
```

```python
len(img_name_vector), train_captions[:5]
```

```python
len(img_name_vector), img_name_vector[:6] # 한개의 사진에 달려있는 캡션이 5개이상
```

```python
from tqdm.notebook import tqdm
from konlpy.tag import Okt

tokenizer = Okt()
token_list = []
for caption in tqdm(train_captions):
    token_list.append(tokenizer.morphs(caption))

token_list_ch = []
for tokens in tqdm(token_list):
    for token in tokens:
        if token not in token_list_ch:
            token_list_ch.append(token)
```

```python
token_dict = dict()
token_dict[0] = 'pad'
token_dict[1] = 'unk'
for index, value in enumerate(token_list_ch):
    token_dict[index+2] = value
print(len(token_dict))
```

```python
index_to_word = token_dict.copy()
word_to_index = dict()
for key, value in index_to_word.items():
    word_to_index[value] = key
word_to_index
```

```python
len(word_to_index) ,len(index_to_word)
```

```python
print(len(train_captions))
preprocessed_sentences = []
for i in tqdm(range(len(train_captions))):
    result = tokenizer.morphs(train_captions[i])
    preprocessed_sentences.append(result)
# print(preprocessed_sentences)
```

```python
# 정수인코딩
encoded_sentences = []
for sentence in preprocessed_sentences:
    encoded_sentence = []
    for word in sentence:
        try:
            encoded_sentence.append(word_to_index[word])
        except KeyError:
            encoded_sentence.append(word_to_index['unk'])
    encoded_sentences.append(encoded_sentence)
# print(encoded_sentences)
```

```python
cap_vector = tf.keras.preprocessing.sequence.pad_sequences(encoded_sentences, padding='post')
```

```python
# Find the maximum length of any caption in the dataset
def calc_max_length(tensor):
    return max(len(t) for t in tensor)
```

```python
# Calculates the max_length, which is used to store the attention weights
max_length = calc_max_length(encoded_sentences)
max_length
```
